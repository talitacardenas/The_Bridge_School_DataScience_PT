# Regresión logística (0 hidden layers neural network)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# tensorflow low level library
# keras high level library
from tensorflow import keras
from tensorflow.keras import models
from tensorflow.keras import layers
```

# ANN - Artificial neural networks

construimos el primer modelo basado en redes neuronales para la predicción del cancer. ¿Mejoramos la precisión de un simple modelo de ML de Clasificación?

```python
ruta_fichero = "https://raw.githubusercontent.com/marcusRB/The_Bridge_School_DataScience_PT/main/04_MachineLearning/00_datasets/breast_cancer.csv"
df = pd.read_csv(ruta_fichero)
df.head()
```

```python
df.describe()
```

```python
x = df.drop("is_cancer", axis=1)
y = df.is_cancer
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2, 
                                                    random_state=17)
```

```python
print(X_train.shape[0])
print(X_test.shape[0])
```

```python
# Realizamos nustro modelo de regres. logística
from sklearn.linear_model import LogisticRegression
```

```python
# Creamos nuestro estimador
log = LogisticRegression(max_iter=10000)
```

```python
# creamos nuestro ajuste
log.fit(X_train, y_train)
```

```python
# Extraemos el score
log.score(X_test, y_test)
```

```python
# Extraemos los coeficientes del modelo
pd.Series(log.coef_[0], index=x.columns).sort_values().tail()
```

La regresión logística busca el mejor *weights* $w_i$, una por feature, y un parámetro extra, el *bias* $b$, para predecir el *outputp* $y$, dadas las features de entrada *inputs* $x$.

Para poder realizar la predicción de una nueva instancia realizamos los siguientes pasos:

$$
z = \displaystyle\sum_{i}x_i \times w_i+b
$$

$$
y = \frac{1}{2+e^-z}
$$

# Multilayer Perceptron (1+ hidden layers neural network)

```python
# creación de una red neuronal
network = models.Sequential()
```

Para construir el core de una red neuronal, debemos crear las capas **layer**, que tiene diferentes nodos **nodes**.

Nuestra red neuronal tendrá:
- **input layer**: dimension 30 (número de predictores)
- **hidden layer**: dimension 10 ?¿
- **output layer**: dimension 1 (is_cancer)

```python
# revisams los inputs
X_train.shape
```

```python
# Añadimos las Hidden Layers
network.add(layers.Dense(10, activation='relu', input_shape=(30,)))

# Añadimos nuestra capa Output
network.add(layers.Dense(1, activation='sigmoid'))
```

Resumiendo:

- durante la fase de entreno, las capas ocultas deben buscar el mejor equilibrio para determinar el *weight*.
- una vez realizada la predicción, de cada feature $x_i$ debemos multiplicar por el peso.
- sigue el proceso de activación.

¿Cuántos pesos necesitamos?

# Compilation de la red neuronal

Para realizar una red neuronal y que sea lista para poder entrenar, necesitamos **compilar** tres aspectos importantes:

- **Loss Function** (métrica diferenciable), así es como la red podrá medir qué tan bien está haciendo un trabajo con sus datos de training.
- **Optimizer** es el mecanismo a través del cual la red se actualizará en función de los datos que ve.
- **Metrics** para monitorear durante la fase de training y el testing. Aquí solo nos preocuparemos por la métrica de **accuracy**.

```python
# realizamos la compilación
network.compile(
    optimizer = 'adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

# Training

Una red neuronal será entrenada en fases **epochs**. En cada step:
1. realiza la predicción a través del cual nos devuelve el *weights* - forward propagation.
2. realiza una comparativa con las etiquetas reales
3. sucesivamente aplica los camios a los pesos *weigths* - back propagation.

```python
# aplicamos a nuestro modelo los parámetros de entreno
network.fit(X_train, y_train.astype(float),
            epochs=3000,
            validation_split=0.1)
# validation_split será el porcentaje de muestras que no serán utilizadas durante la fase 
# de entreno, solo para las epochs en la fase de validación
```

# Resultados

```python
# Predictions
network.predict(X_test)[:20].round(3)
```

```python
# realizamos la comparativa
y_test[:20]
```

```python
# aplicamos la métrica
from sklearn.metrics import accuracy_score
print(f"La precisión de nuestro modelo NN es del {round(accuracy_score(network.predict(X_test) > 0.5, y_test) *100,3)}%")
```
