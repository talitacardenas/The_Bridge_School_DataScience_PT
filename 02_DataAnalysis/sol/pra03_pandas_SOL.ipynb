### Ejercicio 1

Cargad los datos del fichero _got.csv_ en un _dataframe_. Este conjunto de datos recoge información de la [Guerra de los Cinco Reyes](https://awoiaf.westeros.org/index.php/War_of_the_Five_Kings) de las novelas de [Cancion de Hielo y Fuego](https://es.wikipedia.org/wiki/Canci%C3%B3n_de_hielo_y_fuego) de George R.R Martin.

Mostrad el número de filas del _dataframe_ y las etiquetas de los ejes.

```python
# Respuesta
# Importamos la librería `pandas`.
import pandas as pd

# Cargamos los datos del fichero en un dataframe.
df_got = pd.read_csv('https://raw.githubusercontent.com/marcusRB/The_Bridge_School_DataScience_PT/main/02_DataAnalysis/dataset/got.csv', delimiter = ",", header = 0)

# Mostramos el número de filas del dataframe y las etiquetas de los ejes.
print(f"El número de filas del dataframe es {len(df_got)}")
print(f"Las etiquetas de los ejes del dataframe son: \n {df_got.axes}")
```

### Ejercicio 2

Agrupad los datos cargados en el ejercicio 1 por el principal bando atacante (*attacker\_1*). Para cada posición, mostrad el número de batallas y el resultado de la batalla (el resultado se encuentra en el campo *attacker\_outcome*).

```python
# Respuesta
# Agrupamos los datos por el principal bando atacante (attacker_1) y lo guardamos en una variable.
df_got_grouped = df_got.groupby(['attacker_1'])

# Contamos el número de batallas para cada grupo con la función `count`.
df_got_grouped['attacker_outcome'].count()
```

```python
# Mostramos el resultado de las batallas para cada grupo con la función `value_counts`.
df_got_grouped['attacker_outcome'].value_counts()
```

```python
# Iteramos sobre el bando atacante principal agrupado y mostramos la información que nos interesa de cada grupo.
for (attacker, battles) in df_got.groupby(['attacker_1']):
    # Al iterar sobre el groupby pandas devuelve el valor de cada grupo y un dataframe con las líneas que lo forman.
    print(f"El número de batallas que realizaron los {attacker} fueron: {len(battles)}")
```

```python
# Volvemos a iterar, añadiendo información más detallada como es cuántas batallas fueron ganadas por cada grupo.
for (attacker, battles) in df_got.groupby(['attacker_1']):
    print(f"El número de batallas que realizaron los {attacker} fueron {len(battles)} , de las cuales {len(battles[(battles.attacker_outcome == 'win')])} fueron victoriosas")
```

```python
# Respuesta
import numpy as np

# creamos las condiciones y las alternativas
conditions = [
              (df_got.attacker_size > df_got.defender_size),
              (df_got.attacker_size < df_got.defender_size),
              (pd.isna(df_got.attacker_size) & pd.notna(df_got.defender_size)),
              (pd.notna(df_got.attacker_size) & pd.isna(df_got.defender_size))
]

choices = ["True", "False", "False", "True"]
# Creamos la nueva columna con la función `select` de numpy y las condiciones/alternativas anteriormente creadas. Rellenamos
# la columna por defecto con "NaN", que se rellenará en los casos en los que ningún bando tenga tropas, para el resto de
# posibilidades se rellenarán valores "True" o "False" según las condiciones establecidas.
df_got['winner_prediction'] = np.select(conditions, choices, default = "NaN")
```
