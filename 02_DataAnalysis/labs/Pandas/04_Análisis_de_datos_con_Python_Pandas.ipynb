## Formato JSON

`JSON`, acrónimo de _JavaScript Object Notation_, es un formato estándard de transferencia de datos que utiliza una sintaxi legible para los humanos y que consiste en listas anidadas de pares clave/valor. Por [ejemplo](http://json.org/example.html), aquí podemos ver un registro de una base de datos no relacional en formato JSON:

```json
{
  "menu": {
    "id": "file",
    "value": "File",
    "popup": {
      "menuitem": [
        {"value": "New", "onclick": "CreateNewDoc()"},
        {"value": "Open", "onclick": "OpenDoc()"},
        {"value": "Close", "onclick": "CloseDoc()"}
      ]
    }
  }
}
```

Como podemos ver, los datos no están estructurados en formato tabla. Cuando carguemos el `JSON` a un `DataFrame` con `pandas`, cada columna corresponderá a una de las claves de primer nivel de cada registro. Y su valor será, en el caso de listas anidadas, una lista en formato `JSON`.

````python
ruta_fichero = "https://raw.githubusercontent.com/marcusRB/The_Bridge_School_DataScience_PT/main/02_DataAnalysis/dataset/pandas_issues.json"
df_json = pd.read_json(ruta_fichero)
df_json.head()
````

No obstante podemos observar que existen valores en formato diccionario o listas. La forma más cómoda de leer este fichero, es utilizar una normalización de esta tabla. Por lo tanto definimos **aplanar** la estructura JSON

````python
# Para poder aplanar (flatten) un JSON
import json
from pandas.io.json import json_normalize

with open('/content/pandas_issues.json') as f:
  issues = json.load(f)
````

````python
issues
````

````python
type(issues[0])
````

````python
issues[0].items()
````

````python
# Para poder normalizar este diccionario utilizamos json_normalize
df = json_normalize(issues)
df.head()
````

````python
df.info
````

````python
# Para guardar el fichero nuevamente en json
df.to_json("nuevo_fichero.json")
````

### Primera parte de EDA (Exploratory Data Analysis)

- visualizamos los datos
- visualizamos la distribución
- check de valores nulos / vacíos 
- análisis entre variables

````python
# En casp de encontrarnos con filas de valores nulos, eliminaremos la fila entera con dropna
dataset['missings'] = np.nan # creamos una columna para ensayar dropna
````

````python
dataset.info()
````

````python
dataset.head()
````

````python
# Para poder valorar los nulos
'missings' in dataset.columns and dataset['missings'].isnull().all()
````

````python
# Ante de nada guardamos una copia de seguridad
df_bak = dataset
````

````python
# para eliminar los valores nulos cuando estén presentes tanto en fila que columna
dataset.dropna(axis=1,how='all')
````

````python
dataset.head()
````

````python
# Para poder escribir la operación de borrado de los valores nulos es utilziar inplace=True
dataset.dropna(axis=1,how='all', inplace=True) # automaticamente reemplazará el dataset
````

````python
dataset.head(3)
````

````python
'missings' in dataset.columns # comprobamos que la columna missings esté o no 
````

````python
# Operamos con how='any'
dataset.dropna(axis=1, how='any') # comprobamos por columna
````

> la operación anterior eliminaría 21 columnas por estar presente n valores nulos por columna.

````python
# comprobamos por fila los valores nulos
dataset.dropna(axis=0, how='any')
````

````python
dataset.dropna(axis=0, how='all') # con esta operación me aseguro que no hay todos valores nulos
````

````python
# Para poder comprobar a nivel global los valores nulos
dataset.isnull().sum() #por valor absoluto
````

````python
lengthDF = dataset.shape[0]
````

````python
# Para poder comprobar a nivel global los valores nulos
dataset.isnull().sum()/len(dataset)*100 # observamos por valor relativo
````

> Si optamos con el establecer un umbral, por ejemplo, el 5% de los valores nulos

````python
delete_columns = dataset.columns[dataset.isnull().sum()/lengthDF*100 > 5]
delete_columns
````

````python
# Comprobamos la operación
dataset.isnull().sum()/lengthDF*100
````

````python
#si le aplicamos una condición
dataset.isnull().sum()/lengthDF*100 > 5
````

````python
dataset.columns[dataset.isnull().sum()/lengthDF*100 > 5]
````

````python
# eliminaremos las columnas cuando cumple cno nuestro umbral
dataset.drop(delete_columns, axis=1).head() # hay que añadir inplace=True para escribir la operación
````
