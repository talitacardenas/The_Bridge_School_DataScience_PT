En `numpy`el objeto básico se trata de una lista multidimensional de números normalmente del mismo tipo.

```python
# Ejemplo de creación de un array
arreglo = np.array([1,2,3])
arreglo
```

```python
type(arreglo) # comprobamos el tipo de objeto
```

```python
# El array de numpy anterior es del tipo unidimensional, y también podemos extraer los elementos por su índice
arreglo[1]
```

En *NumPy* , a las dimensiones se les conoce con el *nombre de ejes( axes )* y al *número de los ejes rango (rank)*.

Array es un alias para referirse al tipo de objeto *numpy.ndarray*

Tenemos algunas de las propiedades de los **arrays**
- *ndarray.ndim* : el número de ejes del objeto array (matriz)
- *ndarray.shape*: una tupla de números enteros indicando longitud de las dimensiones de la matriz
- *ndarray.size*: número total de elementos de una matriz

```python
# Ejemplo de creación de una matriz 3x2 (3 filas y 2 columnas) - serie unidimensional
matriz = np.arange(3*2)
print("Array unidimensional: ")
print(matriz)
```

```python
type(matriz)
```

```python
# Si quiero darle una forma de matriz 3x2, utilizamos el método `reshape()`
matriz = matriz.reshape(3,2)
matriz # el resultado es una matriz multidimensional de 3 filas y 2 columnas
```

```python
# quiero extraer solo la segunda fila (teniendo en cuenta que la lista empieza por índice 0)
matriz[1]
```

```python
# si quiero extraer los valores de la primera columna
matriz[:,0]
```

```python
# si quiero extraer los valores de la segunda
matriz[:,1]
```

```python
# si quiero extraer el primer valor de la segunda columna (siempre se comienza por índice cero)
matriz[0,1]
```

```python
type(matriz[1,1])
```

```python
# para conocer la dimensión de esta matriz
matriz.ndim
```

```python
# si quiero conocer la forma
matriz.shape
```

```python
# si quiero obtener el número de elementos
matriz.size
```

## Operaciones con matrices

Con *NumPy* podemos implementar todas las operaciones habituales con matrices

```python
# construimos dos matrices
A = np.array([[1,0],[0,1]])
B = np.array([[1,2],[3,4]])
print(A)
print("^"*10)
print("<>"*10)
print(B)
```

```python
# Suma de matrices
print(A+B)
```

```python
# Resta de matrices
print(A-B)
```

```python
# multiplicando matrices elemento por elemento
print(A*B)
```

```python
# multiplicar matriz por matriz
print(A.dot(B))
```

```python
# Potencia en matrices
print(A**2)
```

Otros ejemplos de uso de métodos en NumPy

```python
lst = list(range(1_000_000))
type(lst)
```

```python
%%timeit
for i in lst:
  doubles = []
  q = i**2
  doubles.append(q)
```

```python
import numpy as np
import scipy
from scipy import ndimage
import matplotlib.pyplot as plt

f = scipy.misc.face(gray=True)
sx, sy = f.shape
X, Y = np.ogrid[0:sx, 0:sy]


r = np.hypot(X - sx/2, Y - sy/2)

rbin = (20* r/r.max()).astype(np.int)
radial_mean = ndimage.mean(f, labels=rbin, index=np.arange(1, rbin.max() +1))

plt.figure(figsize=(5, 5))
plt.axes([0, 0, 1, 1])
plt.imshow(rbin, cmap=plt.cm.nipy_spectral)
plt.axis('off')

plt.show()
```
