## A√±adir y quitar datos

```python
# Podemos realizar una carga de datos parciales
df = pd.read_csv("https://raw.githubusercontent.com/stefmolin/Hands-On-Data-Analysis-with-Pandas-2nd-edition/2nd_edition/ch_02/data/earthquakes.csv",
                 usecols=['time','title','place','magType',
                          'mag','alert','type','tsunami'])
df
```

```python
# Podemos crear un nuevo atributo vac√≠o o con valores por defecto
df['source'] = 'USGS API'
```

```python
df.head(10)
```

```python
df.mag < 0
```

```python
# Creamos un nuevo atributo con una condici√≥n del atributo mag < 0
df['mag_negative'] = df.mag < 0
df
```

```python
df[df.mag < 0] # realizamos un subset para visualizar la condici√≥n anterior
```

```python
df[df.mag < 0].describe()
```

```python
# Operamos con las expresiones regulares
df.place.str.extract(r', (.*$)')[0].sort_values().unique()
```

```python
# Operamos con transformaci√≥n de unos valores
df['parsed_place'] = df.place.str.replace(
    r'.* of ', '', regex=True # elimina cualquier elemento que tenga of
).str.replace(
    'the ', '' # eliminamos el the
).str.replace(
    r'CA$', 'California', regex=True
).str.replace(
    r'NV$', 'Nevada', regex=True
).str.replace(
    r'MX$', 'Mexico', regex=True
).str.replace(
    r' region$', '', regex=True
).str.replace(
    'Fiji Islands', 'Fiji'
).str.replace(
    r'^.*, ', '', regex=True # elimina cualquier cosa rara que encuentra al principio del texto
).str.strip() # elimina cualquier espacio
```

```python
df.parsed_place.sort_values().unique()
```

```python
# una v√°lida alternativa es utilizar assign()
df.assign(
    in_ca=df.parsed_place.str.endswith('California'),
    in_alaska=df.parsed_place.str.endswith('Alaska')).sample(5, random_state=0)
```

```python
# Tambi√©n podemos hacer uso de la funci√≥n lambda()
df.assign(
    in_ca=df.parsed_place == 'California',
    in_alaska=df.parsed_place == 'Alaska',
    neither=lambda x: ~x.in_ca & ~x.in_alaska
).sample(5, random_state=0)
```

```python
# Podemos realizar filtros y concatenar
tsunami = df[df.tsunami == 1]
no_tsunami = df[df.tsunami == 0]
tsunami.shape, no_tsunami.shape
```

```python
# Concatenar
pd.concat([tsunami, no_tsunami]).shape
```

```python
# Otra forma de realizarlo es con append()
tsunami.append(no_tsunami).shape
```
