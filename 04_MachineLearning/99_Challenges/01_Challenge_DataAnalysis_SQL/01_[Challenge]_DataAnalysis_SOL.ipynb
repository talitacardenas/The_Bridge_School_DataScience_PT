## Generar fichero n1 - num_creditos, suma financiada

```python
q_task1 = '''
select
	O.merchant_uuid as merchant_id,
	count(O.uuid) as num_creditos,
	round(sum(O.booking)::numeric, 2) as Sales
from public.orders as O
inner join public.merchants as M 
on O.merchant_uuid = M.uuid
where O.number_instalments > 0 and O.annual_percentage_rate > 0
group by merchant_id
order by  num_creditos desc
'''
```

```python
df1 = pd.read_sql_query(sql=q_task1, con=dbConnection)
df1.head(5)
```

```python
df1.set_index(df1['merchant_id'], inplace=True)
df1
```

```python
df1.drop(['merchant_id'], axis=1, inplace=True)
df1
```

## Task02 - cod_sector, sector, num_creditos, ventas

```python
q_task2 = '''
select
	M.industry_code as cod_sector,
	M.industry_name as sector,
	count(O.uuid) as num_creditos,
	round(sum(O.booking)::numeric, 2) as Sales
from public.orders as O
inner join public.merchants as M 
on O.merchant_uuid = M.uuid
where O.number_instalments > 0 and O.annual_percentage_rate > 0
group by sector, cod_sector
order by cod_sector asc
'''
```

```python
df2 = pd.read_sql_query(sql=q_task2, con=dbConnection)
df2.head(5)
```

```python
df2 = df2.groupby(['cod_sector','sector']).sum()
df2
```

```python
df2.to_excel('task2.xlsx', index=True, sheet_name="task2.2")
```

## Task03 - mes, code_sector, sales

```python
q_task3 = '''
select
	to_char(O.created, 'MM') as mes,
	M.industry_code as code_sector,
	round(sum(O.booking)::numeric, 2) as Sales
from public.orders as O
inner join public.merchants as M 
on O.merchant_uuid = M.uuid
where O.number_instalments > 0 and O.annual_percentage_rate > 0
group by 1,2
order by mes asc, code_sector asc
'''
```

```python
df3 = pd.read_sql_query(sql=q_task3, con=dbConnection)
df3.head(5)
```

```python
df3 = pd.pivot_table(df3, values = 'sales', index=['mes'], columns = 'code_sector')
df3
```

```python
df3.to_excel('task3.xlsx', index=True, sheet_name="task2.3")
```
